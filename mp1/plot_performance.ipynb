```python
import matplotlib.pyplot as plt

# Data for plotting
categories = [
    "4 Machines Frequent", "4 Machines Infrequent", "4 Machines count", "4 Machine No match",
    "10 Machines Frequent", "10 Machines Infrequent", "10 Machines count", "10 Machine No match"
]
avg_values = [1.7352, 0.3632, 0.0664, 0.052, 3.8558, 0.7448, 0.07, 0.0584]
std_values = [0.1352800798, 0.03387772129, 0.001140175425, 0.003391164992,
              0.05033090502, 0.04701808163, 0.004, 0.005683308895]
std_errors = [
    0.0604990909, 0.01515057755, 0.0005099019514, 0.001516575089,
    0.022508665, 0.02102712534, 0.001788854382, 0.002541653005
]

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(categories, avg_values, label='Average', color='blue', s=10)
plt.errorbar(categories, avg_values, yerr=std_values, fmt='o', color='red', capsize=5, label='Std Deviation')

# Annotate average and standard deviation values on the plot
for i, (avg, std) in enumerate(zip(avg_values, std_values)):
    plt.text(i, avg + 0.02, f"Avg: {avg:.4f}\nStd: {std:.4f}", ha='center', va='bottom')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Scatter Plot of Averages with Standard Deviation Error Bars')
plt.xticks(rotation=45, ha='right')
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
