```python
import numpy as np
import matplotlib.pyplot as plt

# Given arrays
array1 = [1234, 1321, 1102]
array2 = [1534, 1343, 1421]

# Compute average (mean) and standard deviation
mean1 = np.mean(array1)
std_dev1 = np.std(array1)

mean2 = np.mean(array2)
std_dev2 = np.std(array2)

# Data for plotting
labels = ['RainStorm', 'SparkStreaming']

means = [mean1, mean2]
std_devs = [std_dev1, std_dev2]

# Plotting
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x, means, width, label='process time', yerr=std_devs, capsize=2, color=['orange'])

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('System')
ax.set_ylabel('Time (s)')
ax.set_title('Process time of RainStorm and SparkStreaming')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
