```python
import numpy as np
import matplotlib.pyplot as plt

# Given arrays
array1 = [0.87, 0.76, 0.79, 0.93, 0.85]
array2 = [0.46, 0.53, 0.50, 0.51, 0.41]
array3 = [0.21, 0.23, 0.22, 0.21, 0.21]
array4 = [0.1, 0.09, 0.12, 0.1, 0.1]
array5 = [0.017, 0.034, 0.023, 0.036, 0.015]



# Compute average (mean) and standard deviation
mean1 = np.mean(array1)
std_dev1 = np.std(array1)

mean2 = np.mean(array2)
std_dev2 = np.std(array2)

mean3 = np.mean(array3)
std_dev3 = np.std(array3)

mean4 = np.mean(array4)
std_dev4 = np.std(array4)

mean5 = np.mean(array5)
std_dev5 = np.std(array5)

# Data for plotting
labels = ['100MB', '50MB', '20MB', '5MB', '1MB']

means = [mean1, mean2, mean3, mean4, mean5]
std_devs = [std_dev1, std_dev2, std_dev3, std_dev4, std_dev5]

# Plotting
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x, means, width, label='Bandwidth', yerr=std_devs, capsize=5, color=['orange'])

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('File Size')
ax.set_ylabel('second')
ax.set_title('Re-replication Time')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Attach a text label above each bar in *rects*, displaying its height and standard deviation.
def autolabel(rects, std_devs):
    """Attach a text label above each bar in *rects*, displaying its height and standard deviation."""
    for rect, std_dev in zip(rects, std_devs):
        height = rect.get_height()
        ax.annotate(f'{round(height, 2)}\nÂ±{round(std_dev, 2)}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# autolabel(rects1, std_devs)

fig.tight_layout()

plt.show()
```
